```python
import os
import pandas as pd
import numpy as np
```

```python
train_dir = os.listdir('dataset-sign-new/train')
train_dir
```

```python
# Delete all images in train dir that has 'rotate' in their name
for dir in train_dir:
	dir_path = os.path.join('dataset-sign-new/train', dir)
	images = os.listdir(dir_path)
	for image in images:
		if 'rotate' in image:
			print('Deleting', os.path.join(dir_path, image))
			os.remove(os.path.join(dir_path, image))
```

```python
import shutil
import random

for dir in train_dir:
	dir_path = os.path.join('dataset-sign-new/train', dir)
	images = os.listdir(dir_path)
	num_val = len(images) // 10 # Split 10% of the data to val
	val_dir = os.path.join('dataset-sign-new/val', dir)
	os.makedirs(val_dir, exist_ok=True)
	random_images = random.sample(images, num_val)
	print('Moving', num_val, 'images from', dir_path, 'to', val_dir)
	for img in random_images:
		img_path = os.path.join(dir_path, img)
		shutil.move(img_path, val_dir)
```
